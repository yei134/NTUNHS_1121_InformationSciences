# 資料處理

```python
# 將類別型物件資料進行轉換
data["ShelveLoc"]=data["ShelveLoc"].astype("category")
data["Urban"]=data["Urban"].astype("category")
data["US"]=data["US"].astype("category")
```

```python
data["ShelveLoc"]
```

```python
new_ShelveLoc = [0,2,1]
data["code_ShelveLoc"] = data["ShelveLoc"].cat.rename_categories(new_ShelveLoc)
data["code_ShelveLoc"]
```

```python
data["Urban"]
```

```python
new_Urban = [0,1]
data["code_Urban"] = data["Urban"].cat.rename_categories(new_Urban)
data["code_Urban"]
```

```python
data["US"]
```

```python
new_US = [0,1]
data["code_US"] = data["US"].cat.rename_categories(new_US)
data["code_US"]
```

# 檢視資料

```python
"""
Sales- 每個地點的銷售量（以千為單位）
.CompPrice- 競爭對手在每個地點收取的價格
Income- 社區收入水準（單位：千美元）
Advertising- 公司在每個地點的本地廣告預算（以千美元為單位）
Population- 地區人口規模（以千計）
.Price- 每個地點的汽車座椅價格公司收費
.ShelveLoc- 具有“不良”、“良好”和“中等”級別的因素，指示每個站點的汽車座椅擱架位置的質量
Age-當地人口的平均年齡
Education-每個地點的教育水平
Urban- 具有“否”和“是”級別的因素，指示商店位於城市還是農村地區
US- 具有“否”和“是”級別的因素，指示商店是否位於美國
"""
#用散佈圖方式顯示資料
sns.relplot(x="CompPrice",y="Price",data=data,hue="ShelveLoc")
# 透過顯示可以看出廠商價格與競爭廠商價格呈現正向相關
# 但是品質與價格沒有明顯分區與相關
```

# 作業

```python
"""
.Sales- 每個地點的銷售量（以千為單位）
CompPrice- 競爭對手在每個地點收取的價格
Income- 社區收入水準（單位：千美元）
Advertising- 公司在每個地點的本地廣告預算（以千美元為單位）
Population- 地區人口規模（以千計）
.Price- 每個地點的汽車座椅價格公司收費
ShelveLoc- 具有“不良”、“良好”和“中等”級別的因素，指示每個站點的汽車座椅擱架位置的質量
Age-當地人口的平均年齡
Education-每個地點的教育水平
Urban- 具有“否”和“是”級別的因素，指示商店位於城市還是農村地區
US- 具有“否”和“是”級別的因素，指示商店是否位於美國
"""
# 是不是不好的 shelveLoc (Bad) 增加了廣告預算後 (Advertising)，銷售就會變好 (Sales, CompPrice, Price... 合成一個值，或依照權重比例合成一個值)
# Ans:銷售額不會因為廣告增加而接受品質不好的產品
# 好的 shelvLoc (Good) 不用增加廣告預算，銷售就會好?
# Ans:
data["total_sales"]=data["Sales"]*data["Price"]
ShelveLoc_bad = data[data["ShelveLoc"]=="Bad"]
ShelveLoc_good = data[data["Advertising"]<5]
# ShelveLoc_good = ShelveLoc_good[ShelveLoc_good["Advertising"]<5]
sns.relplot(x="Advertising",y="total_sales",data=ShelveLoc_bad,hue="ShelveLoc")
sns.relplot(x="Advertising",y="Price",data=ShelveLoc_bad,hue="ShelveLoc")
sns.relplot(x="Advertising",y="Sales",data=ShelveLoc_bad,hue="ShelveLoc")
sns.relplot(x="Advertising",y="total_sales",data=ShelveLoc_good,hue="ShelveLoc")
sns.relplot(x="Advertising",y="Price",data=ShelveLoc_good,hue="ShelveLoc")
sns.relplot(x="Advertising",y="Sales",data=ShelveLoc_good,hue="ShelveLoc")
sns.relplot(x="Advertising",y="total_sales",data=data,hue="ShelveLoc")
sns.relplot(x="Advertising",y="Sales",data=data,hue="ShelveLoc")
sns.relplot(x="Advertising",y="Price",data=data,hue="ShelveLoc")
```

```python
"""
Sales- 每個地點的銷售量（以千為單位）
CompPrice- 競爭對手在每個地點收取的價格
Income- 社區收入水準（單位：千美元）
Advertising- 公司在每個地點的本地廣告預算（以千美元為單位）
Population- 地區人口規模（以千計）
Price- 每個地點的汽車座椅價格公司收費
ShelveLoc- 具有“不良”、“良好”和“中等”級別的因素，指示每個站點的汽車座椅擱架位置的質量
Age-當地人口的平均年齡
Education-每個地點的教育水平
Urban- 具有“否”和“是”級別的因素，指示商店位於城市還是農村地區
US- 具有“否”和“是”級別的因素，指示商店是否位於美國
"""
# 好的銷售跟那些欄位有關係?  => Population?Education.... 正相關?
tmp = data[["Sales","CompPrice","Income","Advertising","Population","Price","code_ShelveLoc","Age","Education","code_Urban","code_US"]]
# 相關性分析
tmp = tmp.corr()
tmp
```

```python
sns.heatmap(tmp,center=0.7)
```
